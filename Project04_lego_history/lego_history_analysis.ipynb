## 1. What percentage of all licensed sets ever released were Star Wars themed?

```python
sets.shape
```

```python
themes.shape
```

```python
sets.isna( ).sum( )
```

```python
sets = sets[~sets["set_num"].isna( )]
sets.isna( ).sum( )
```

```python
themes.isna( ).sum( )
```

```python
themes[themes["is_licensed"] == True]
```

```python
df = sets.merge( themes, left_on = "parent_theme", right_on = "name" )
df
```

```python
df.drop( columns = "name_y", inplace = True )
df
```

```python
df.columns = df.columns.str.replace( "name_x", "name" )
df
```

```python
licensed_df = df[df["is_licensed"] == True]
licensed_df
```

```python
star_wars = licensed_df[licensed_df["parent_theme"] == "Star Wars"]
percentage = int( np.round( 100 * star_wars.shape[0] / licensed_df.shape[0] ) )
print( str( percentage ) + "%" )
```

### Approximately 52 percent of all licensed sets were Star Wars themed, which is more than a half of all licensed sets.

## 2. In which year was Star Wars not the most popular licensed theme (in terms of number of sets released that year)?

```python
licensed_df
```

```python
licensed_df["count"] = 1
```

```python
grouped = licensed_df.groupby( ["year", "parent_theme"] )[["count"]].sum( ).reset_index( )
grouped
```

```python
sorted_no_duplicates = grouped.sort_values( by = ["year", "count"], ascending = False ).drop_duplicates( "year" )
sorted_no_duplicates
```

```python
year = sorted_no_duplicates[sorted_no_duplicates["parent_theme"] != "Star Wars"]["year"]
year
```

### It can be seen above that the only year when Star Wars was NOT the most popular licensed theme is 2017, when Super Heroes managed to take over Star Wars. All other years starting from 1999 to 2016 Star Wars was the most popular theme.

## 3. How many different sets were released each year? (not necessarily licensed)

```python
df
```

```python
df["count"] = 1
df.pivot_table( index = "year", values = "count", aggfunc = "sum" ).plot( figsize = ( 15, 9 ), kind = "bar" )
plt.show( )
```

### It can be seen that over the years, starting from 1950s, the number of sets being released each year increases. For instance, in 1997 there were around 180 sets released, while in 1998 - more than 300 sets.
### The greatest number of sets were released in the year of 2014, where the number was approximately 700.
