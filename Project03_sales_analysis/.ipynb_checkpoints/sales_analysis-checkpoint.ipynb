```python
import glob # in order to read multiple CSV files from the directory
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# get a list of CSV files from the directory
path = "./data"
csv_files = glob.glob( path + "/*.csv" )

# read each CSV file into a DataFrame
df_list = ( pd.read_csv( file ) for file in csv_files )
df = pd.concat( df_list, ignore_index = True )

# Combine the resultant DataFrame to a CSV file that consists of all the data
df.to_csv( "concatenated_data.csv", index = False )
```

```python
df = pd.read_csv( "concatenated_data.csv" )
df
```

```python
df[df["Product"].isna( )]
```

#### The columns should be converted to their appropriate data types.

```python
pd.options.mode.chained_assignment = None
df["Quantity Ordered"] = df["Quantity Ordered"].astype( "int32" )
df["Price Each"] = df["Price Each"].astype( "float64" )
```

#### So there are some issues with the rows of the DataFrame.

```python
df[df["Quantity Ordered"].str[:] == "Quantity Ordered"]
```

#### These rows certainly need to be deleted.

```python
df = df[~( df["Quantity Ordered"].str[:] == "Quantity Ordered") ]
```

```python
# Converting to appropriate data types
df["Quantity Ordered"] = df["Quantity Ordered"].astype( "int32" )
df["Price Each"] = df["Price Each"].astype( "float64" )
```

```python
df["Order Date"] = df["Order Date"].apply( lambda x : pd.to_datetime( x ) )
```

```python
df.info( )
```

```python
df
```

```python
# Add the month column to the DataFrame
df["Month"] = df["Order Date"].dt.month
df
```

#### Let's first calculate the total cost of the purchase by multiplying "Quantity Ordered" column by "Price Each".

```python
df["Total Sales"] = df["Quantity Ordered"] * df["Price Each"]
df
```

## 1. What was the best month for sales? How much was earned that month?

```python
month_vs_sales = pd.pivot_table( df, index = "Month", values = "Total Sales", aggfunc = "sum" )
month_vs_sales
```

```python
month_vs_sales.plot( figsize = ( 15, 9 ), kind = "bar", rot = 0 )
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun", "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]
plt.xticks( np.arange( 12 ), months )
plt.xlabel( "Month" )
plt.ylabel( "Sales in USD ($)" )
plt.show( )
```

### The best month for sales was December, when the company earned slightly above 4600000 USD. Why did that happen? One possible reason is because of Christmas or New Year when a lot of people buy gifts.

```python
df[df["Month"] == 12].groupby( "Product" )[["Quantity Ordered"]].agg( "count" )
```

### As it can be seen, the most popular product to be ordered in December was USB-C Charging Cable with 2980 orders. It is followed by Lightning Charging Cable (2893 orders) and Batteries of type AAA (2831 orders)

## 3. What time should the advertisement be displayed to maximize the likelihood of customers' buying products? 

```python
df["Hour"] = df["Order Date"].dt.hour
df["Minute"] = df["Order Date"].dt.minute
df
```

```python
hour_vs_sales = pd.pivot_table( df, index = "Hour", values = "Total Sales", aggfunc = "count" )

hour_vs_sales.plot( figsize = ( 12, 8 ) )
plt.xticks( np.arange( 24 ), np.arange( 0, 24, 1 ) )
plt.xlabel( "Hour" )
plt.ylabel( "Number of orders" )
plt.grid( )
plt.show( )
```

### There are peaks of number of sales at 12am and 7pm, which implies that these are the best moments of people making orders. So, if the ad was posted at approximately 10-11am and at 4-5pm, the number of sales could be even more maximized.

## 4. What products are most often sold together? (two products are sold together if their Order ID is the same)

```python
# For instance
df.head( 20 )
```

### Rows 3 and 4 have different products but the same Order ID, which implies that these goods were sold together.

```python
# Creates a new DataFrame that only keeps rows with duplicated Order ID column.
df_with_duplicates = df[df["Order ID"].duplicated( keep = False )]
df_with_duplicates
```

```python
df_with_duplicates["Grouped"] = df.groupby( "Order ID" )["Product"].transform( lambda x : ','.join( x ) )
df_with_duplicates
```

```python
df_with_duplicates = df_with_duplicates[["Order ID", "Grouped"]].drop_duplicates( )
df_with_duplicates
```

```python
from itertools import combinations
from collections import Counter

count = Counter( )

for row in df_with_duplicates["Grouped"]:
    row_lst = row.split( ',' )
    count.update( Counter( combinations( row_lst, 2 ) ) ) # 2 means the combinations of 2 ordered items: among all products,
                                                          # the pair of iPhone and Lightning Charging Cable were the most
                                                          # often sold together. The most often sold triple of products is
                                                          # Google Phone, USB-C Charging Cable, and Wired Headphones.

count.most_common( 10 )
```

## 5. What product was sold the most? And why it was sold the most?

```python
product_vs_number = pd.pivot_table( df, index = "Product", values = "Quantity Ordered", aggfunc = "sum" )
product_vs_number
```

### AAA Batteries were sold the most in the year of 2019. This might be due to its high demand as it is used in most of the electronic devices as well as due to its availability. Batteries are not as much expensive as laptops or monitors, so they are bought more often.

```python
# So I need to check the hypothesis - the difference in the price of the most and least sold products.
product_vs_price = pd.pivot_table( df, index = "Product", values = "Price Each", aggfunc = "mean" )
product_vs_price
```

```python
fig, ax1 = plt.subplots( figsize = ( 15, 9 ) )

ax2 = ax1.twinx( )
ax1.bar( product_vs_number.index, product_vs_number["Quantity Ordered"], color = 'g' )
ax2.plot( product_vs_price.index, product_vs_price["Price Each"], color = 'b' )

ax1.set_xlabel( "Product name" )
ax1.set_xticklabels( labels = product_vs_price.index, rotation = 90 )
ax1.set_ylabel( "Number ordered", color = 'g' )
ax2.set_ylabel( "Unit price in USD ($)", color = 'b' )

plt.show( )
```

### So it can be seen that the higher the unit price of a product, the less often it is ordered by customers and vice versa. Therefore, the hypothesis that AAA Batteries are the most sold product due to their affordability (unit price is only 2.99 USD).
