```python
df[df["ceiling_height"] < ceiling_mean - ceiling_std]
```

## Работа с типами данных

```python
# Следующие столбцы могут быть переведены в тип данных int, поскольку они обозначают целочисленные характеристики:
# не может быть 2,5 комнаты или 19,3-го этажа
df[["total_images", "rooms", "floor", "balcony",
    "parks_around3000", "ponds_around3000", "days_exposition"]] = df[["total_images", "rooms", "floor", "balcony",
    "parks_around3000", "ponds_around3000", "days_exposition"]].astype( "int64" )
```

```python
df[["last_price", "total_area", "ceiling_height"]] = df[["last_price", "total_area", "ceiling_height"]].astype( "float" )
```

## Добавление новых столбцов в таблицу и работа с аномальными данными.

```python
df["cost/m2"] = df["last_price"] / df["total_area"] # Цена на квадратный метр
df["day_name"] = df["first_day_exposition"].dt.day_name( ) # Получить название дня недели
df["day_number"] = df["first_day_exposition"].dt.day_of_week # Получить индекс дня недели (Monday=0, Sunday=6)
df["month"] = df["first_day_exposition"].dt.month # Получить месяц из столбца "first_day_exposition"
df["year"] = df["first_day_exposition"].dt.year # Получить год из столбца "first_day_exposition"

def get_floor( row ):
    if row["floor"] == 1:
        return "Первый"
    elif row["floor"] == row["floors_total"]:
        return "Последний"
    else:
        return "Другой"

df["floor_type"] = df.apply( get_floor, axis = 1 )
```

```python
# Квартиры, цена за квадратный метр которых превышает одно стандартное отклонение.
df[( df["cost/m2"] > df["cost/m2"].mean( ) + df["cost/m2"].std( ) )][["last_price", "total_area", "locality_name", "cost/m2"]]
```

```python
df["cost/m2"].describe( )
```

#### Минимальная стоимость одного квадратного метра квартиры составляет 9615, а максимальная превышает 1 миллион. Эти аномальные значения необходимо исправить.
#### Можно использовать только среднее значение и значения, находящиеся в интервале одного стандартного отклонения, чтобы стоимость одного квадратного метра подчинялась нормальному распределению.

```python
# Группировка данных по наименованию города, чтобы определить среднее и стандартное отклонение отдельно для каждого города
# и тем самым избавиться от аномальных значений для каждого города отдельно.
def remove_cost_outliers( df ):
    df_out = pd.DataFrame( )
    for key, subdf in df.groupby( "locality_name" ):
        cost_mean = np.mean( subdf["cost/m2"] ) # средняя цена квадратного метра
        cost_std = np.std( subdf["cost/m2"] ) # стандартное отклонение
        reduced_df = subdf[( subdf["cost/m2"] >= ( cost_mean - cost_std ) ) & ( subdf["cost/m2"] <= ( cost_mean + cost_std ) )]
        df_out = pd.concat( [df_out, reduced_df], ignore_index = True )
    return df_out

df = remove_cost_outliers( df )
df["cost/m2"].describe( )
```

```python
plt.figure( figsize = ( 15, 10 ) )
plt.hist( df["cost/m2"], bins = 20 )
plt.title( "Cost per square meter Histogram" )
plt.xlabel( "Cost per square meter, ₽/m²" )
plt.xticks( np.arange( 0, 180000, 10000 ) )
plt.ylabel( "Number of observations" )
plt.grid( )
plt.show( )
```

#### По гистограмме выше видно, что стоимость квадратного метра квартиры подчиняется нормальному распределению и что большинство пользователей продают квартиры по цене в промежутке от 70000 до 120000 рублей за квадратный метр. Есть так же и "аномальные" значения - менее 30000 и более 150000 рублей за м². Вероятно, эти значения принадлежат квартирам в посёлках или деревнях, отдаленных от центра и аэропорта и жилым домам в самом центре, соответственно.

```python
# Группировка данных по площади недвижимости и подсчет средней цены квартир с данной площадью.
# Сгруппированные данные хранятся в сводной таблице, которая создается при помощи функции pivot_table( ).
price_vs_area = pd.pivot_table( df, index = "total_area", values = "last_price", aggfunc = "mean" )

plt.figure( figsize = ( 15, 9 ) )
# price_vs_area.index содержит сгруппированные площади квартир.
plt.scatter( price_vs_area.index, price_vs_area["last_price"], label = "Mean Price vs. Total Area" )
plt.xlabel( "Total Area, m²" )
plt.ylabel( "Real estate price, ₽" )
plt.legend( ) # Чтобы отобразить label: ("Mean Price vs. Total Area")
plt.show( )
```

#### Безусловно, цена квартиры сильно зависит от общей площади, но это не единственный параметр. По графику выше видно, что у квартиры с большей площадью цена может совпадать, а то и быть ниже, чем у квартиры с меньшей площадью. Возможно, это потому что на цену влияют еще такие параметры, как удаленность от центра города или аэропорта, число комнат, этажность и т.д.

```python
price_vs_rooms = pd.pivot_table( df, index = "rooms", values = "last_price", aggfunc = "mean" )

plt.figure( figsize = ( 15, 9 ) )
plt.bar( price_vs_rooms.index, height = price_vs_rooms["last_price"], label = "Mean Price vs. Number of Rooms" )
plt.xticks( np.arange( 1, 20, 1 ) )
plt.xlabel( "Number of Rooms" )
plt.ylabel( "Real estate price, ₽" )
plt.legend( )
plt.show( )
```

#### Диаграмма выше демонстрирует, что с увеличением количества комнат в квартире, также возрастает и средняя цена недвижимости. Однако, цена падает после 9 и 15-комнатных квартир. Возможно, тут играют роль такие параметры как удаленность от центра города или аэропорта или количество парков и водоёмов в радиусе 3км.

#### Казалось бы, чем дальше недвижимость от центра города, тем цена должна быть дешевле, но точечная диаграмма выше показывает, что есть квартиры, находящиеся очень близко к центру города и стоящие дешевле чем другие, находящиеся подальше от центра.

#### 10 населенных пунктов с наибольшим числом объявлений.

```python
# Группировка данных датасета по названию города.
# Метод "count" считает, сколько раз встречается каждый город в датасете. Далее идет сортировка данных по убыванию,
# чтобы вывести города с наибольшим числом объявлений в начале.
ten_most_cities = df.groupby( "locality_name" )["locality_name"].agg( "count" ).sort_values( ascending = False )[:10]
ten_most_cities
```

#### Средняя цена квадратного метра в этих населенных пунктах.

```python
# Группировка данных по городу, подсчет количества повторений каждого города и среднего значения цены за квадратный метр
# в каждом городе.
most_mean_cost_per_square_m = pd.pivot_table( df, index = "locality_name", values = "cost/m2", aggfunc = ["count", "mean"] )

# Сортировка количества повторений каждого города по убыванию, чтобы наиболее частые города были впереди.
# [:10] означает, что выбираются только 10 значений.
ten_most_df = most_mean_cost_per_square_m.sort_values( by = ( "count", "cost/m2" ), ascending = False )["mean"][:10]
ten_most_df
```

```python
# Группировка данных по городу, подсчет количества повторений каждого города, минимального и максимального значений
# общей цены недвижимости в каждом городе.
max_min_cities = pd.pivot_table( df, index = "locality_name", values = "last_price", aggfunc = ["count", "max", "min"] )
max_min_cities = max_min_cities.sort_values( by = ( "count", "last_price" ), ascending = False )[:10]

# Выбор названия населенного пункта с максимальной и минимальной стоимостью жилья.
# .index возвращает название, .values[0] конвертирует в тип данных str, чтобы вывести в предложении ниже.
max_city = max_min_cities[max_min_cities["max"]["last_price"] == max_min_cities["max"]["last_price"].max( )].index.values[0]
min_city = max_min_cities[max_min_cities["min"]["last_price"] == max_min_cities["min"]["last_price"].min( )].index.values[0]
```

```python
print( "Населенный пункт с самой высокой стоимостью жилья:", max_city )

print( "Населенный пункт с самой низкой стоимостью жилья:", min_city )
```

#### 10 населенных пунктов с наименьшим числом объявлений.

```python
# Группировка данных датасета по названию города.
# Метод "count" считает, сколько раз встречается каждый город в датасете. Далее идет сортировка данных по возрастанию,
# чтобы вывести города с наибольшим числом объявлений в начале.
ten_least_cities = df.groupby( "locality_name" )["locality_name"].agg( "count" ).sort_values( ascending = True )[:10]
ten_least_cities
```

```python
least_mean_cost_per_square_m = pd.pivot_table( df, index = "locality_name", values = "cost/m2", aggfunc = ["count", "mean"] )
ten_least_df = least_mean_cost_per_square_m.sort_values( by = ( "count", "cost/m2" ), ascending = True )["mean"][:10]
ten_least_df
```

```python
max_min_cities = pd.pivot_table( df, index = "locality_name", values = "last_price", aggfunc = ["count", "max", "min"] )
max_min_cities = max_min_cities.sort_values( by = ( "count", "last_price" ), ascending = True )[:10]
max_city = max_min_cities[max_min_cities["max"]["last_price"] == max_min_cities["max"]["last_price"].max( )].index.values[0]
min_city = max_min_cities[max_min_cities["min"]["last_price"] == max_min_cities["min"]["last_price"].min( )].index.values[0]
```

```python
print( "Населенный пункт с самой высокой стоимостью жилья:", max_city )

print( "Населенный пункт с самой низкой стоимостью жилья:", min_city )
```

```python
# Конкатенация двух датасетов вместе: 10 населенных пунктов с наибольшим и 10 с наименьшим числом объявлений.
ten_most_and_least = pd.concat( [ten_most_df, ten_least_df] )
ten_most_and_least
```
