# Data Cleaning

### It is considered that such columns as "Developer Website", "Developer Email", "Privacy Policy", "Editors Choice", and "Scraped Time" do not affect the analysis of the dataset and cannot give any insight about the given data. That is why these columns can be dropped.

```python
df.drop( columns = ["Developer Website", "Developer Email", "Privacy Policy", "Editors Choice", "Scraped Time"],
         inplace = True )
df.info( )
```

### Moreover, it is seen that the columns "Installs" and "Minimum Installs" contain duplicate values, so one of the columns can be dropped as well.

```python
df.drop( columns = "Minimum Installs", inplace = True )
df
```

### Let's check for NaN values in the columns.

```python
# These NaN values should not affect the analysis since there is a column named "App Id", which represents a unique ID of
# each app.
df[df["App Name"].isna( )]
```

```python
df[df["Rating"].isna( )]
```

### If there is info about the ratings or rating counts of apps, then maybe these apps did not get any ratings from the users, so both columns can be set to 0.

```python
df.loc[df["Rating"].isna( ), ["Rating", "Rating Count"]] = 0
df.isna( ).sum( )
```

```python
df[df["Installs"].isna( )]
```

### These NaN values of the "Installs" columns can be easily replaced with correct values based on the "Maximum Installs" column values. That is, if there were 67 or more installs, I will replace NaN with 50+, if 3000 or more, I will replace NaN with 1000+, etc.

```python
def set_installs( row ):
    max_installs = str( row["Maximum Installs"] )
    if int( max_installs[0] ) >= 5:
        return str( 5 * 10 ** ( len( max_installs ) - 1 ) ) + "+"
    elif int( max_installs[0] ) >= 1:
        return str( 1 * 10 ** ( len( max_installs ) - 1 ) ) + "+"
    else:
        return "0+"


df["Installs"] = df.apply( set_installs, axis = 1 )

df.iloc[[5622, 43569, 60888, 64322, 91391, 2161180]]
```

```python
df[df["Currency"].isna( )]
```

```python
df[( df["Currency"].isna( ) ) & ( df["Price"] > 0 )]
```

```python
df[( df["Currency"] == "XXX" ) & ( df["Price"] > 0 )]
```

### It can be seen that there are no rows that have NaN currency and price more than 0 (paid apps). It means that these apps are free, therefore the currency does not apply to these apps - so it can be set to XXX.

```python
df.fillna( value = { "Currency" : "XXX" }, inplace = True )
df.isna( ).sum( )
```

```python
df[( df["Size"].str.contains( "Var" ) ) & ( df["Size"].isna( ) == False )]
```

### Rows above contain data about apps in Google Play whose size "Varies with device". Both these and NaN values need to be dropped because they do not allow for a better data analysis. That is, I will not be able to compare the size of the app with its number of installs by users.

```python
df[df["Size"].isna( )]
```

```python
df = df[~( df["Size"].isna( ) )]
df = df[~( df["Size"].str.contains( "Var" ) )]
df.isna( ).sum( )
```

```python
df[( df["Minimum Android"].str.contains( "Var" ) ) & ( df["Minimum Android"].isna( ) == False )]
```

### The same above method can be applied to "Varies with device" and NaN values in the "Minimum Android" column.

```python
df = df[~( df["Minimum Android"].isna( ) )]
df = df[~( df["Minimum Android"].str.contains( "Var" ) )]
df.isna( ).sum( )
```

```python
df[df["Released"].isna( )]
```

### It can be assumed that the released and last updated dates coincide, so NaN values in the former column can be replaced with the date specified in the "Last Updated" column. This may be the case if these apps were not updated since the release.

```python
df[df["Released"] == df["Last Updated"]]
```

### It is evident from here, that there are many application (more than 300000) that were not updated since the release.

```python
def fill_released( row ):
    return row["Last Updated"]


df["Released"] = df.apply( fill_released, axis = 1 )
df.isna( ).sum( )
```

```python
df[df["Developer Id"].isna( )]
```

### There are only 32 rows that contain NaN values in the "Developer Id" columns, so these can be dropped.

```python
df.dropna( subset = ["Developer Id"], inplace = True )
df.info( )
```

```python
df.isna( ).sum( )
```

```python
df.head( )
```

### "Minimum Android" column has redundant information by saying "and up", so I can get rid of that.

```python
def replace_android( row ):
    return row["Minimum Android"].split( ' ' )[0]


df["Minimum Android"] = df.apply( replace_android, axis = 1 )
df
```

# Checking for outlier values and features engineering

### First, let's check if there are apps that are free - column "Free" is True - but the value in the column "Price" is not 0 (or vice versa).

```python
df[( df["Free"] == True ) & ( df["Price"] > 0 )]
```

```python
df[( df["Free"] == False ) & ( df["Price"] == 0 )]
```

### These rows contradict each other: they are not free but their price is 0, so they should be deleted.

```python
df = df[~( ( df["Free"] == False ) & ( df["Price"] == 0 ) )]
```

### Checking the "Size" column.

### Let's check for the values of the "Size" column. First, I need to convert them to a uniform size (maybe to MegaBytes or KiloBytes).

```python
df[~df["Size"].str.contains( "M" )]
```

```python
df[~( df["Size"].str.contains( "M" ) ) & ~( df["Size"].str.contains( "k" ) )]
```

### k near the number in the "Size" column may imply kilobytes, whereas G - gigabytes. I will convert all the sizes to Megabytes.

```python
# In case these rows contain commas to separate 1000's.
df[df["Size"].str.contains( "," )]
```

```python
df.loc[df["Size"].str.contains( ',' ), "Size"] = df["Size"].str.replace( ',', '' )
```

```python
def convert_size( row ):
    if 'k' in row["Size"]:
        return float( row["Size"][:-1] ) / 1024
    elif 'M' in row["Size"]:
        return float( row["Size"][:-1] )
    elif 'G' in row["Size"]:
        return float( row["Size"][:-1] ) * 1024
    

df["Size"] = df.apply( convert_size, axis = 1 )
type( df["Size"][0] )
```

```python
df.info( )
```
